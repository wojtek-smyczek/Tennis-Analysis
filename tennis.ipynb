```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib.ticker as ticker # 

pd.set_option('display.max_columns', None)
file_path = 'atp_matches_till_2022.csv'

df = pd.read_csv(file_path)

print("Data loaded with success")


df.sample(5)
```

```python
year = pd.to_numeric(df['tourney_id'].str[:4], errors='coerce')

df_filtered = df[year >= 1992]


print(f"Original rows: {len(df)}")
print(f"Filtered rows: {len(df_filtered)}")

df_filtered.head()
```

We are taking the number of wins for every player and taking the top10

```python
total_wins = df_filtered['winner_name'].value_counts().head(20)

print(" Top 10 total matches won from 1992 ")
print(total_wins)

```

```python
data_to_plot = total_wins.sort_values(ascending=True)

plt.figure(figsize=(10,7))
data_to_plot.plot(kind='barh') #barh - wykres slupokowy

plt.title('Most wins in ATP from 1992 to 2022')
plt.xlabel('Number of wins')
plt.ylabel('PLayer')

plt.show()
```

```python
sns.set_style("whitegrid")

df['tourney_date'] = pd.to_datetime(df['tourney_date'], format='%Y%m%d', errors='coerce')
df['year'] = df['tourney_date'].dt.year # tworzymy kolumne year
df = df.dropna(subset=['year']) # usuwamy kolumny gdzie year jest pusta
df['year'] = df['year'].astype(int) 


# filtrujemy dane
big_3 = ['Roger Federer', 'Novak Djokovic', 'Rafael Nadal']
df_big3 = df[df['winner_name'].isin(big_3) & (df['year'] >= 1992)]


yearly_wins = df_big3.groupby(['year', 'winner_name']).size().reset_index(name='wins')


yearly_wins = yearly_wins.sort_values(by=['winner_name', 'year'])
yearly_wins['cumulative_wins'] = yearly_wins.groupby('winner_name')['wins'].cumsum()


plt.figure(figsize=(12, 7))


for player in big_3:
    player_data = yearly_wins[yearly_wins['winner_name'] == player]
    plt.plot(player_data['year'], player_data['cumulative_wins'], label=player, linewidth=2.5)
    



plt.title('cumulative number of wins( 1992 - 2022)', fontsize=16)
plt.xlabel('Rok', fontsize=12)
plt.ylabel('Łączna liczba wygranych meczów', fontsize=12)
plt.legend(title='Player', fontsize=10)


plt.ylim(bottom=0)

plt.tight_layout() # Dopasowuje wykres, aby nic nie zostało ucięte



plt.show()


```


Most wins in the Grand Slams.

```python
df_slams = df_filtered[df_filtered['tourney_level'] == 'G']
slams_win = df_slams['winner_name'].value_counts().head(10)

slam_data_to_plot = slams_win.sort_values(ascending=True)

plt.figure(figsize=(10,7))
slam_data_to_plot.plot(kind='barh', color='green')

plt.title('Most grand slams wins 1992 - 2022')

plt.show()



```

Checking when the Wildcard player has won the whole tournamnet.

```python
cond_final = df_filtered['round'] == 'F'

cond_wildcard = df_filtered['winner_entry'] == 'WC'

cond_top10 = df_filtered['loser_rank'] <= 10

combined_cond = df_filtered[cond_final & cond_wildcard & cond_top10]

cols_to_show = ['tourney_date', 
                'tourney_name', 
                'surface', 
                'winner_name', 
                'winner_rank', 
                'loser_name', 
                'loser_rank']

# Zastosuj listę kolumn i posortuj
final_results = combined_cond[cols_to_show].sort_values(by='tourney_date', ascending=False)

print("Finals in which the player with the Wildcard defeated the Top 10 player:")
final_results
```

Histogram dla czasu trwania meczu (min)

```python
# delete the NaN values in 'min' cols
df_minutes = df.dropna(subset=['minutes'])

plt.figure(figsize=(10, 6))
sns.histplot(df_minutes['minutes'], bins=200, kde=True, color='blue')

plt.title('Length of ATP matches (minutes)', fontsize=15)
plt.xlabel('Length of a single match', fontsize=12)
plt.ylabel('Frequency of matches', fontsize=12 )

plt.xlim(30,300)

# getting the axis to modify it
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(10)) # podzialka co 10 jednostek



plt.tight_layout()
plt.show()
```

Corelation between height of a player and number of aces served in a match.

```python


sns.set_style("whitegrid")


# Create a table with winner performances
df_winners = df[['winner_ht', 'w_ace']].dropna()
df_winners.columns = ['height', 'aces'] 

# Create a table with loser performances
df_losers = df[['loser_ht', 'l_ace']].dropna()
df_losers.columns = ['height', 'aces'] 


df_performance = pd.concat([df_winners, df_losers])

df_performance_clean = df_performance[df_performance['height'] > 150]

print(f"We will analyze {len(df_performance_clean)} individual player performances.")

# Obliczamy korelacje
correlation = df_performance_clean['height'].corr(df_performance_clean['aces'])

print(f"\nThe correlation coefficient between height and aces is: {correlation:.4f}")


print("Generating correlation plot...")
#  'alpha=0.1' for transprant dots
g = sns.jointplot(
    data=df_performance_clean, 
    x='height', 
    y='aces', 
    kind='reg', # 'reg' = scatterplot + regression line
    height=8,
    joint_kws={'scatter_kws': {'alpha': 0.1, 's': 15}}, # 's' is the point size
    line_kws={'color': 'red', 'linewidth': 2} # Style of the trend line
)

g.set_axis_labels('Player Height (cm)', 'Aces in Match', fontsize=12)
g.fig.suptitle(f'Correlation between Height and Aces (r = {correlation:.2f})', y=1.03, fontsize=16)


plt.tight_layout()
plt.savefig('correlation_height_aces.png')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
df_height = df.dropna(subset=['winner_ht', 'surface'])

df_height = df_height[df_height['winner_ht'] > 150] 

sns.boxplot(data=df_height, x='surface', y='winner_ht', order=['Clay', 'Hard', 'Grass', 'Carpet'])
plt.title('Distribution of Winner Height by Surface')
plt.xlabel('Surface')
plt.ylabel('Winner Height (cm)')
plt.show()
```

```python

# Calculate the total number of aces (winner and loser) in each match, filling NaNs with 0
df['total_aces'] = df['w_ace'].fillna(0) + df['l_ace'].fillna(0)


yearly_avg_aces = df.groupby('year')['total_aces'].mean()


rolling_avg_aces = yearly_avg_aces.rolling(window=5, center=True).mean()


plt.figure(figsize=(12, 6))
plt.plot(yearly_avg_aces.index, yearly_avg_aces.values, label='Yearly Average', alpha=0.5)
plt.plot(rolling_avg_aces.index, rolling_avg_aces.values, label='5-Year Rolling Average', color='red', linewidth=2)


plt.title('Trend: Average Aces per Match on ATP Tour (Since 1992)')
plt.xlabel('Year')
plt.ylabel('Average Aces per Match')
plt.legend() 


plt.show()
```
